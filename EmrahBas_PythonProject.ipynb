```python
# Ülkelere göre ödül sayısını hesapla
country_counts = data['birth_country'].value_counts()

# En çok ödül kazanan ülkenin adını belirle
top_country = country_counts.idxmax()

# Amerika'nın aldığı ödüllerin olduğu yılları belirle
usa_years = data[data['birth_country'] == top_country]['year'].unique()

# Amerika'nın aldığı ödüllerin olduğu yıllarda diğer ülkelerin aldığı ödül sayısını hesapla
other_countries_awards_in_usa_years = data[data['year'].isin(usa_years) & (data['birth_country'] != top_country)]['year'].value_counts().sort_index()

# Amerika'nın aldığı ödül sayısını hesapla
usa_awards_by_year = data[data['birth_country'] == top_country]['year'].value_counts().sort_index()

# Hakimiyet grafiğini çiz
plt.figure(figsize=(10, 6))

# Diğer ülkelerin aldığı ödül sayısını göster
plt.bar(other_countries_awards_in_usa_years.index, other_countries_awards_in_usa_years.values, color='salmon', label='Diğer Ülkeler')

# Amerika'nın aldığı ödül sayısını göster
plt.bar(usa_awards_by_year.index, usa_awards_by_year.values, color='skyblue', label=top_country)

plt.title(f"{top_country}'nin Ödül Aldığı Yıllarda Diğer Ülkeler ve {top_country}'nin Aldığı Ödül Sayısı")
plt.xlabel("Yıl")
plt.ylabel("Ödül Sayısı")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=70)
plt.xticks(np.arange(min(usa_years), max(usa_years)+1, 2))
plt.tight_layout();
plt.show()
```

#### 5. Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz. (Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.
### Cevap: 

```python
# Her ödül kategorisi için ayrı görselleştirme yapmak için kategori listesi
categories = data['category'].unique()

# Bir alt grafik dizisi oluştur
fig, axes = plt.subplots(len(categories), 3, figsize=(15, len(categories) * 5))

# Her bir kategori için döngü
for i, category in enumerate(categories):
    # Belirli kategoriye göre verileri filtrele
    category_data = data[data['category'] == category]
    
    # Cinsiyet dağılımı
    sns.countplot(data=category_data, x='sex', ax=axes[i, 0])
    axes[i, 0].set_title(f'{category} - Cinsiyet Dağılımı')
    axes[i, 0].set_xlabel('Cinsiyet')
    axes[i, 0].set_ylabel('Count')
    
    # Yaş sütunundaki eksik değerleri dikkate almadan yaş dağılımı (KDE)
    category_data_without_nan = category_data.dropna(subset=['age'])
    sns.histplot(data=category_data_without_nan, x='age', bins=20, kde=True, ax=axes[i, 1])
    axes[i, 1].set_title(f'{category} - Yaş Dağılımı')
    axes[i, 1].set_xlabel('Yaş')
    axes[i, 1].set_ylabel('Count')
    
    # Ödül yılı dağılımı
    sns.histplot(data=category_data, x='year', bins=20, kde=True, ax=axes[i, 2])
    axes[i, 2].set_title(f'{category} - Ödül Yılı Dağılımı')
    axes[i, 2].set_xlabel('Yıl')
    axes[i, 2].set_ylabel('Count')

# Alt grafikler arasındaki boşlukları ayarla
plt.tight_layout()
plt.show()

```

#### 7. 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız. (Her kategori için ayrı bir grafik olması istenmektedir)
### Cevap:

```python
# Belirli zaman aralığındaki verileri filtreleyelim
filtered_data = data[(data['year'] > 1947) & (data['year'] < 1991)]

# Ödül kategorilerini alalım
prize_categories = filtered_data['category'].unique()

# Her bir ödül kategorisi için ayrı bir grafik oluşturalım
for category in prize_categories:
    # Ödül kategorisine göre veriyi filtreleyelim
    category_data = filtered_data[filtered_data['category'] == category]
    
    # Yeni bir figür oluşturalım
    plt.figure(figsize=(5, 5))
    
    # Kategoriye göre kazananların ülke dağılımını çizelim
    sns.countplot(y='birth_country', data=category_data, order=category_data['birth_country'].value_counts().index)
    plt.title(f'{category} Kategorisinde Nobel Ödülü Kazananların Ülkeleri (1947-1991)')
    plt.xlabel('Kişi Sayısı')
    plt.ylabel('Ülke')
    
    # Grafikleri gösterelim
    plt.show()
```

#### 8.Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin. (Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.
### Cevap:

```python
# Belirli kategorileri seçme (soruda ekonomi kategorisi yok)
selected_categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']

# Her bir kategori için döngü oluşturma
for category in selected_categories:
    # Belirli kategorideki verileri filtreleme
    category_data = data[(data['category'] == category) & (data['year'] > 2000)]
    
    # Yeni bir figür oluşturma
    fig, axes = plt.subplots(1, 2, figsize=(20, 6))
    
    # Ülke dağılımını çizme
    sns.countplot(y='birth_country', data=category_data, order=category_data['birth_country'].value_counts().index, ax=axes[0])
    
    axes[0].set_title(f'{category} Kategorisindeki Nobel Ödülü Kazananların Ülkeleri (2000 Sonrası)')
    axes[0].set_xlabel('Kişi Sayısı')
    axes[0].set_ylabel('Ülke')
    
    # Yaş dağılımını çizme
    sns.histplot(data=category_data, x='age', bins=20, kde=True, ax=axes[1])
    
    axes[1].set_title(f'{category} Kategorisindeki Nobel Ödülü Kazananların Yaş Dağılımı (2000 Sonrası)')
    axes[1].set_xlabel('Yaş')
    axes[1].set_ylabel('Kişi Sayısı')
    
    # Grafikleri gösterme
    plt.show()

```

Ağırlıklı olarak ABD ve Avrupa ülkelerinden kazananlar var. Yaş dağılımı genellikle 50-70 yaş arasında, bu da önemli fiziksel keşiflerin genellikle kariyerin orta dönemlerinde gerçekleştiğini gösteriyor.

#### Tıp: 
Kazananların çoğu ABD, Avrupa ve Japonya'dan. Yaş dağılımı 40-80 yaş arasında değişiyor, bu da tıbbi araştırmaların geniş bir kariyer yelpazesinde tanınabileceğini gösteriyor.

###### Bu görselleştirmeler, Nobel Ödülleri'nin genellikle kariyerlerinin orta veya ileri dönemlerindeki bireylere verildiğini, ancak özellikle Barış ödülünde genç kazananların da olduğunu göstermektedir. Coğrafi dağılım açısından, ABD ve Avrupa'nın hala hakim olduğu, ancak özellikle Edebiyat ve Barış kategorilerinde dünya genelinden kazananların olduğu görülmektedir.
